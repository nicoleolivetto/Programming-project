```python
topsubscribed_df = topsubscribed_df.assign(Years_on_YT = 2023 - topsubscribed_df['Started'])
topsubscribed_df = topsubscribed_df.assign(Ten_Years="0")
```

```python
df2 = topsubscribed_df.drop('Category', axis=1)
df2 = df2.drop('Youtube Channel', axis=1)
df2 = df2.drop('range_subs', axis=1)
df2.reset_index(drop=True, inplace=True)
df2
```

```python
q = df2["Video Count"].quantile(0.75)
q_low = df2["Video Count"].quantile(0.25)
q_hi  = df2["Video Count"].quantile(0.75)

df_filtered = df2[(df2["Video Count"] < q_hi) & (df2["Video Count"] > q_low)]
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

```python
plt.figure(figsize=(8,6))
sb.heatmap(df_filtered.corr(), annot=True)
```

```python
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

```python
for i in range (len(df_filtered['Ten_Years'])):
    if((df_filtered.at[(i), 'Years_on_YT']) > 10):
        df_filtered.at[(i), 'Ten_Years'] = 1
    else:
        df_filtered.at[(i), 'Ten_Years'] = 0
```

```python
df_filtered
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.model_selection  import train_test_split
```

```python
X = df_filtered[['Subscribers', 'Rank']]
Y = df_filtered['Video Views']
Y=Y.astype('int')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_filtered, Y,train_size=0.8, test_size=0.2, random_state=42)
from sklearn.metrics import mean_absolute_error as mae
```

```python
from sklearn.linear_model import LinearRegression , Lasso
ln = LinearRegression()
ln.fit(X_train,y_train)
pred = ln.predict(X_test)
print('mae = ',mae(y_test,pred))
```

```python
ls = Lasso(0.2)
ls.fit(X_train,y_train)
lasso_pred = ls.predict(X_test)
mae(y_test,lasso_pred)
     
```

```python

#random forrest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
rf_pred = rf.predict(X_test)
mae(y_test,rf_pred)
     
```

```python

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler
x_norm = MinMaxScaler().fit_transform(df_filtered)
chi_selector = SelectKBest(chi2, k=3)
chi_selector.fit(x_norm, Y)
chi_support = chi_selector.get_support()
chi_feature = df_filtered.loc[:,chi_support].columns.tolist()
print('selected features=',chi_feature)
```

```python
Xs = df_filtered[chi_feature]
Xs
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xs, Y,train_size=0.8, test_size=0.2,random_state=42)
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression , Lasso
ln = LinearRegression()
ln.fit(X_train,y_train)
pred = ln.predict(X_test)
print('mae = ',mae(y_test,pred))
```

```python
#lasso
ls = Lasso(0.2)
ls.fit(X_train,y_train)
lasso_pred = ls.predict(X_test)
mae(y_test,lasso_pred)
```

```python
#random forrest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
rf_pred = rf.predict(X_test)
mae(y_test,rf_pred)
```
