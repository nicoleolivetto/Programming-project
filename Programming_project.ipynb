```python
#cambio ora con fascia oraria
for i in range(len(strisciate_df['ora_visita'])):
    attribute_column=strisciate_df.at[(i),'ora_visita'].hour
    if (attribute_column >= 8) and (attribute_column <= 12):
        attribute_column = '8am - 12pm'
        strisciate_df.at[(i), 'ora_visita']=attribute_column
    elif (attribute_column > 12) and (attribute_column <= 16):
        attribute_column = '12apm - 4pm'
        strisciate_df.at[(i), 'ora_visita']=attribute_column
    elif (attribute_column > 16) and (attribute_column <= 20):
        attribute_column = '4pm - 8pm'
        strisciate_df.at[(i), 'ora_visita']=attribute_column
    elif (attribute_column > 20) and (attribute_column <= 24):
        attribute_column = '48pm - 12am'
        strisciate_df.at[(i), 'ora_visita']=attribute_column
```
